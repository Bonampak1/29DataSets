### A topographical map with a info blurb as you highlight over the area

```python
#first we need to select our latitude and longitude range
#selecting our latitude range
loc_df_1 = df[(df["latitude"] >= 35) & (df["latitude"] <= 42)]

#selecting our longitude range
loc_df_2 = loc_df_1[(loc_df_1["longitude"] >= -114) & (loc_df_1["longitude"] <= -108)]


loc_df_2.head()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
#with 52 million rows I needed to adjust to just get visualizations, something to consider for final project
sample_size = 1000  

# Get a random sample
df_sample = loc_df_2.sample(n=sample_size, random_state=42)  # Set a random_state for reproducibility


print(loc_df_2.size)
#create a new column with the Total
df_sample['Total'] = df_sample['Crime_Category'].groupby(df_sample['Crime_Category']).transform('count')

#do this on the original
loc_df_2['Total'] = loc_df_2['Crime_Category'].groupby(loc_df_2['Crime_Category']).transform('count')

df_sample.head()

```

```python
import geopandas as gpd
import folium
from folium import plugins
import contextily as cx

# Assuming df_sample is your DataFrame with avalanche data

# get geometry point data for avalanche locations using latitude and longitude
crime_df = gpd.GeoDataFrame(df_sample, geometry=gpd.points_from_xy(df_sample['longitude'], df_sample['latitude']), crs='EPSG:4326')

# convert the GeoDataFrame to a Folium map
m = folium.Map(location=[crime_df['latitude'].mean(), crime_df['longitude'].mean()], zoom_start=10, tiles='OpenTopoMap', attr="OpenTopoMap")

# Define a color mapping for day_of_week
color_mapping = {
    'Sunday': 'blue',
    'Monday': 'green',
    'Tuesday': 'red',
    'Wednesday': 'yellow',
    'Thursday':'orange',
    'Friday': 'black',
    'Saturday': 'purple'
    # Add more days and corresponding colors as needed
}

# Create a legend for the color mapping
legend_html = """
     <div style="position: fixed; 
                 bottom: 190px; left: 150px; width: 150px; height: 180px; 
                 border:2px solid grey; z-index:9999; font-size:14px;
                 background-color:white; opacity: 0.9;
                 ">
     &nbsp; <b>Legend</b> <br>
     &nbsp; Sunday &nbsp; <i class="fa fa-circle fa-1x" style="color:blue"></i><br>
     &nbsp; Monday &nbsp; <i class="fa fa-circle fa-1x" style="color:green"></i><br>
     &nbsp; Tuesday &nbsp; <i class="fa fa-circle fa-1x" style="color:red"></i><br>
     &nbsp; Wednesday &nbsp; <i class="fa fa-circle fa-1x" style="color:yellow"></i><br>
     &nbsp; Thursday &nbsp; <i class="fa fa-circle fa-1x" style="color:orange"></i><br>
     &nbsp; Friday &nbsp; <i class="fa fa-circle fa-1x" style="color:black"></i><br>
     &nbsp; Saturday &nbsp; <i class="fa fa-circle fa-1x" style="color:purple"></i><br>
      </div>
     """

# Add the legend to the map
m.get_root().html.add_child(folium.Element(legend_html))

# Add markers with pop-up blurbs and color them based on day_of_week
for index, row in crime_df.iterrows():
    day_of_week = row['day_of_week']
    color = color_mapping.get(day_of_week, 'gray')  # Default to gray if the day is not in the mapping
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Total: {row['Total']}<br>City: {row['city']}<br>Hour of Day: {row['hour_of_day']}",
        tooltip=row['Crime_Category'],
        icon=folium.Icon(color=color, icon_size=(30, 30))  # Adjust the icon_size parameter to make the markers smaller
    ).add_to(m)

# Add OpenStreetMap tiles as a basemap (which includes streets)
folium.TileLayer('openstreetmap').add_to(m)

# display the map
m.save("crime_data_mapping.html")  # Save the map to an HTML file and open it in a browser
```
